```python
# Drop categorical values
num_feats = data.drop(['id', 'name', 'host_name', 'neighbourhood_group', 'neighbourhood', 'room_type', 'last_review'],
                     axis=1)
num_feats
```

```python
#Replace NA with zeros
df = num_feats.fillna(0)
#num_feats.head()
```

```python
num_feats.dtypes
```

```python
feats = df.drop(['price'], axis=1)
```

```python
#X = df['id', 'host_id', 'latitude', 'longitude', 'minimum_nights', 'number_of_reviews ', 'reviews_per_month'
#             'calculated_host_listings_count', 'availability_365']

X = feats

y = df.price

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=1)

linreg = LinearRegression()

linreg.fit(X_train, y_train)

linreg.coef_

scores = cross_val_score(linreg, X_train, y_train, cv=200, scoring='neg_mean_squared_error')

agg = np.mean(np.sqrt(-scores))

print('Coefficents:', linreg.coef_)

print('Cross validation score:', scores)

print('RMSE:', agg)
```
